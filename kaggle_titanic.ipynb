```python
import pandas as pd
```

```python
teste = pd.read_csv('test.csv')
id_passageiro = teste['PassengerId']
teste
```

```python
treino = pd.read_csv('train.csv')
treino
```

```python
mode_embarked = treino['Embarked'].mode()[0]
treino['Embarked'].fillna(mode_embarked, inplace=True)
```

```python
treino['Embarked'].unique()
```

```python
cabine = treino['Cabin']
treino = treino.drop('Cabin', axis= 1)
ticket = treino['Ticket']
treino = treino.drop('Ticket', axis= 1)
nome = treino['Name']
treino = treino.drop('Name', axis= 1)
x = treino.drop('Survived', axis= 1)
y = treino['Survived']
```

```python
teste = teste.drop('Cabin', axis= 1)
teste = teste.drop('Ticket', axis= 1)
teste = teste.drop('Name', axis= 1)
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
colunas_teste = teste.columns
colunas = x.columns
colunas_teste
```

```python
treino.info()
```

```python
one_hot = make_column_transformer((
    OneHotEncoder(),
    ['Sex', 'Embarked', 'Pclass']
),
    remainder= 'passthrough',
    sparse_threshold= 0)
```

```python
teste = one_hot.fit_transform(teste)
```

```python
teste.shape
```

```python
x = one_hot.fit_transform(x)
```

```python
one_hot.get_feature_names_out(colunas)
```

```python
one_hot.get_feature_names_out(colunas_teste)
```

```python
dt_x = pd.DataFrame(x, columns= one_hot.get_feature_names_out(colunas))
```

```python
dt_x
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
y = label_encoder.fit_transform(y)
```

## Ajustando o modelo Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore = DecisionTreeClassifier(max_depth= 3)
arvore.fit(x, y)
```

```python
arvore.score(x, y)
```

```python
arvore.score(teste, dados['Survived'])
```

```python
previsao = arvore.predict(teste)
```

```python
prev = pd.DataFrame(previsao, columns= ['Survived'])
```

```python
res = pd.concat([id_passageiro, prev], axis= 1)
```
